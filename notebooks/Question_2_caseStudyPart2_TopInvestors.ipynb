```python
# Open and read data file as specified in the question
# Print the required output in given format
import pandas as pd
import numpy as np
from collections import OrderedDict
import matplotlib.pyplot as plt
file=pd.read_csv(r"C:\Users\Dell\Desktop\startup_funding.csv")
df=file.copy()
investors=[]
for row in df.InvestorsName:
    if row != "undisclosed" or row != "Undisclosed":
         l1=str(row).split(",")
         if len(l1)>1:
            for i in l1:
                investors.append(i.strip())
         else:
            investors.append(l1[0].strip())
dict_freq={}
for i in set(investors):
    dict_freq[i]=investors.count(i)
sorted_dict=OrderedDict(sorted(dict_freq.items(),key=lambda x:x[1],reverse=True))
count=0
investors_5=[]
number_5=[]
print("Top 5 INVESTORS","      ","Investments")
print("--------------------------------------")
for key,value in sorted_dict.items():
    if count<5:
        investors_5.append(key)
        number_5.append(value)
        print(key,"-->", value)
    count+=1
explodes=(0.1,0,0,0,0)
plt.pie(number_5, explode=explodes, labels=investors_5, autopct='%1.1f%%',
        shadow=True, startangle=90) 
plt.axis("equal")
plt.show()

    

```
